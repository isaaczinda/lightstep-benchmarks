```python
untraced = []
traced = []

TARGET_SPS = 0
ACTUAL_SPS = 1
CPU_USAGE = 2

with open('untraced.txt') as file:
    while True:
        line = file.readline()
        if not line:
            break
        
        untraced.append([float(i) for i in line.split()])

untraced_sps = [u[ACTUAL_SPS] for u in untraced]
untraced_cpu = [u[CPU_USAGE] for u in untraced]

with open('traced.txt') as file:
    while True:
        line = file.readline()
        if not line:
            break
        
        traced.append([float(i) for i in line.split()])

traced_sps = [u[ACTUAL_SPS] for u in traced]
traced_cpu = [u[CPU_USAGE] for u in traced]
```

# Determining if Normal
Below, we will do an analysis to determine if the data stored in untraced array is normal. If it is, we can safely use mean and stderr to calculate CPU usage for a given slice. 

```python
import matplotlib.pyplot as plt

plt.scatter(untraced_sps, untraced_cpu)
plt.scatter(traced_sps, traced_cpu)

plt.show()
```

```python
untraced_0_sps = list(map(lambda x : x[CPU_USAGE], filter(lambda x : x[TARGET_SPS] == 0, untraced)))
traced_0_sps = list(map(lambda x : x[CPU_USAGE], filter(lambda x : x[TARGET_SPS] == 0, traced)))

# min(untraced_0_sps + traced_0_sps)

bins = numpy.linspace(min(untraced_0_sps + traced_0_sps), max(untraced_0_sps + traced_0_sps), 10)

n, bins, patches = plt.hist(x=untraced_0_sps, bins=bins, color='#0504aa',
                            alpha=0.7, rwidth=0.85)

n, bins, patches = plt.hist(x=traced_0_sps, bins=bins, color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.show()
```

```python
untraced = []
traced = []

SPS = 0
CPU = 1

with open('traced.txt') as file:
    while True:
        line = file.readline()
        if not line:
            break
        
        traced.append([float(i) for i in line.split()])

with open('untraced.txt') as file:
    while True:
        line = file.readline()
        if not line:
            break
        
        untraced.append([float(i) for i in line.split()])
```
